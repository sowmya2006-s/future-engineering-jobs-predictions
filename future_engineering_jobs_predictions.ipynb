```python

import pandas as pd
import matplotlib as mt
from sklearn.linear_model import LinearRegression
df=pd.read_csv("DATA sdc.csv")
mt.pyplot.scatter(df['YEAR'],df['INDEX'],color='red')
mt.pyplot.xlabel('YEAR')
mt.pyplot.ylabel('INDEX')
mt.pyplot.title('future engineering jobs')
```

```python
reg_model=LinearRegression()
reg_model.fit(df[['YEAR']],df.INDEX)
reg_model.coef_
reg_model.intercept_
reg_model.predict([[15]])
```
